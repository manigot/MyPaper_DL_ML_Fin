```
import math
import pandas as pd


def rationalQuadratic(src: pd.Series, lookback: int, relativeWeight: float, startAtBar: int):
    """
    vectorized calculate for rational quadratic curve
    :param src:
    :param lookback:
    :param relativeWeight:
    :param startAtBar:
    :return:
    """
    currentWeight = [0.0]*len(src)
    cumulativeWeight = 0.0
    for i in range(startAtBar + 2):
        y = src.shift(i, fill_value=0.0)
        w = (1 + (i ** 2 / (lookback ** 2 * 2 * relativeWeight))) ** -relativeWeight
        currentWeight += y.values * w
        cumulativeWeight += w
    val = currentWeight / cumulativeWeight
    val[:startAtBar + 1] = 0.0
    return val


def gaussianCurve(src: pd.Series, lookback: int, startAtBar: int):
    """
    vectorized calculate for gaussian curve
    :param src:
    :param lookback:
    :param startAtBar:
    :return:
    """
    currentWeight = [0.0]*len(src)
    cumulativeWeight = 0.0
    for i in range(startAtBar + 2):
        y = src.shift(i, fill_value=0.0)
        w = math.exp(-(i ** 2) / (2 * lookback ** 2))
        currentWeight += y.values * w
        cumulativeWeight += w
    val = currentWeight / cumulativeWeight
    val[:startAtBar + 1] = 0.0
    return val
```

```
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.mixture import GaussianMixture
from scipy.stats import entropy
from torch.utils.data import Dataset, DataLoader, TensorDataset
import torch.nn.functional as F
from sklearn.preprocessing import StandardScaler

class AdvancedFeatureEngineer:
    def __init__(self, df):
        self.df = df.copy()
        self._validate_inputs()

    def _validate_inputs(self):
        required = ['open', 'high', 'low', 'close', 'volume']
        if not all(col in self.df.columns for col in required):
            raise ValueError("DataFrame must contain OHLCV columns")

    # 1. Order Flow Imbalance Dynamics
    def microstructure_features(self, entropy_window=50):
        """Captures market microstructure dynamics"""
        # Microstructure noise ratio
        self.df['microstructure_noise'] = (self.df['high'] - self.df['low']) / (self.df['volume'] + 1e-5)

        # Volume entropy
        def rolling_entropy(s):
            return s.rolling(entropy_window).apply(
                lambda x: entropy(x.value_counts(normalize=True)))

        self.df['volume_entropy'] = rolling_entropy(self.df['volume'])

        # Aggressive flow
        self.df['aggressive_flow'] = (self.df['close'].diff() * self.df['volume']).cumsum()

        return self

    # 2. Auction Market Theory Features
    def auction_features(self, profile_window=30, val_ratio=0.7):
        """Implements auction market theory concepts"""
        def volume_profile(s):
            poc = s.value_counts().idxmax()
            val = s.quantile(val_ratio)
            vah = s.quantile(1-val_ratio)
            return poc, val, vah

        self.df['poc'] = self.df['close'].rolling(profile_window).apply(
            lambda x: volume_profile(x)[0])
        self.df['val'] = self.df['close'].rolling(profile_window).apply(
            lambda x: volume_profile(x)[1])
        self.df['vah'] = self.df['close'].rolling(profile_window).apply(
            lambda x: volume_profile(x)[2])

        # Value area penetration
        self.df['value_penetration'] = (self.df['close'] - self.df['poc']) / (
            self.df['vah'] - self.df['val'] + 1e-5)

        # Auction rejections
        self.df['auction_rejection'] = (
            (self.df['high'] > self.df['vah']) &
            (self.df['close'] < self.df['vah'])
        ).astype(int)

        return self

    # 3. ML-Generated Features (Autoencoder)
    class MarketEncoder(nn.Module):
        """Neural market state encoder"""
        def __init__(self, input_dim=10, latent_dim=3):
            super().__init__()
            self.encoder = nn.Sequential(
                nn.Linear(input_dim, 5),
                nn.Tanh(),
                nn.Linear(5, latent_dim))

        def forward(self, x):
            return self.encoder(x)

    def generate_latent_features(self, feature_columns, latent_dim=3):
        """Create compressed market state representations"""
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        model = self.MarketEncoder(len(feature_columns), latent_dim).to(device)

        # Train on normalized features
        scaler = StandardScaler()
        features = scaler.fit_transform(self.df[feature_columns])
        dataset = TensorDataset(torch.Tensor(features))
        loader = DataLoader(dataset, batch_size=32, shuffle=True)

        # Training loop
        optimizer = torch.optim.Adam(model.parameters())
        for epoch in range(100):
            for batch in loader:
                x = batch[0].to(device)
                z = model(x)
                loss = torch.mean(z**2)  # Basic reconstruction loss
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

        # Generate latent features
        with torch.no_grad():
            latent = model(torch.Tensor(features).to(device)).cpu().numpy()

        for i in range(latent_dim):
            self.df[f'latent_{i}'] = latent[:,i]

        return self

    # 4. Temporal Fractal Dimension
    def fractal_features(self, hurst_window=100, max_lag=50):
        """Calculates fractal market dimension with robust NaN handling"""
        def hurst(series):
            # Validation checks
            if len(series) < max_lag*2 or series.isnull().any():
                return np.nan

            # Remove leading/lagging NaNs
            series = series.dropna()
            if len(series) < max_lag*2:
                return np.nan

            # Check for constant series
            if np.allclose(series, series.iloc[0]):
                return 0.5  # Random walk fallback

            lags = range(2, max_lag)
            tau = []
            valid_lags = []

            # Calculate variance ratios
            for lag in lags:
                if 2*lag > len(series):
                    continue

                diff = series.diff(lag).dropna()
                if diff.var() == 0:
                    continue

                tau.append(np.log(np.sqrt(diff.var())))
                valid_lags.append(lag)

            # Require minimum 5 points for regression
            if len(tau) < 5:
                return np.nan

            # Robust linear regression
            try:
                H, _ = np.polyfit(np.log(valid_lags), tau, 1)
                return H
            except:
                return np.nan

        # Calculate with minimum 50% data
        self.df['hurst'] = self.df['close'].rolling(
            hurst_window, min_periods=int(hurst_window*0.5)
        ).apply(hurst, raw=False)

        # Fill remaining NaNs and calculate entropy
        self.df['hurst'] = self.df['hurst'].fillna(0.5)  # Assume random walk
        self.df['fractal_entropy'] = np.exp(-self.df['hurst'])
        self.df['fractal_entropy'] = self.df['fractal_entropy'].replace(
            [np.inf, -np.inf], 1.0
        )

        return self

    # 5. Liquidity Zone Dynamics
    def liquidity_features(self, lq_window=200, bins=10):
        """Identifies liquidity concentration zones"""
        def liquidity_density(series):
            if len(series) < lq_window:  # Handle initial NaN periods
                return np.nan

            counts, edges = np.histogram(series, bins=bins)

            # Handle edge cases
            last_price = series[-1]
            bin_idx = np.digitize(last_price, edges, right=False) - 1

            # Clamp to valid indices
            bin_idx = max(0, min(bin_idx, len(counts)-1))

            return counts[bin_idx] / (counts.max() + 1e-8)  # Prevent division by zero

        self.df['liquidity_density'] = self.df['close'].rolling(
            lq_window, min_periods=lq_window//2
        ).apply(liquidity_density, raw=True)

        return self

    # 6. Market Regime Classification
    def regime_features(self, regime_window=50, n_components=3):
        """Properly aligned regime classification with edge handling"""
        # Calculate required features
        self.df['volatility'] = self.df['close'].pct_change().rolling(20, min_periods=1).std()

        # Prepare feature matrix with NaN handling
        features = ['volatility', 'volume', 'Kernel_diff']
        X = self.df[features].rolling(regime_window, min_periods=1).mean()
        X = X.ffill().bfill()

        # Create alignment mask (only predict where full window available)
        mask = X.notna().all(axis=1)
        X_valid = X[mask]

        # Fit GMM only on valid data
        gmm = GaussianMixture(n_components=n_components)
        regimes = gmm.fit_predict(X_valid)

        # Create full-length series with original index
        self.df['regime'] = np.nan
        self.df.loc[mask, 'regime'] = regimes

        # Forward fill initial NaNs and convert to category
        self.df['regime'] = self.df['regime'].ffill().fillna(-1).astype('category')

        return self

    # 7. Microstructural Noise Ratio
    def noise_features(self, noise_window=20):
        """Quantifies market noise level"""
        tr = self.df['high'] - self.df['low']
        dm = abs(self.df['close'] - self.df['open'])
        self.df['noise_ratio'] = ((tr - dm)/tr).ewm(span=noise_window).mean()
        return self

    # 8. Temporal Pattern Embeddings
    class PatternCNN(nn.Module):
        """Learns temporal pattern embeddings with proper imports"""
        def __init__(self, input_channels=1):
            super().__init__()
            self.conv1 = nn.Conv1d(input_channels, 8, 5)
            self.conv2 = nn.Conv1d(8, 4, 3)
            self.pool = nn.AdaptiveAvgPool1d(1)

        def forward(self, x):
            x = F.relu(self.conv1(x))
            x = F.max_pool1d(x, 2)
            x = F.relu(self.conv2(x))
            return self.pool(x).squeeze()

    def pattern_features(self, pattern_window=60):
        """Generates properly dimensioned pattern embeddings"""
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        model = self.PatternCNN().to(device)

        # Create valid sequences
        sequences = []
        valid_indices = []
        for i in range(pattern_window, len(self.df)):
            seq = self.df['close'].iloc[i-pattern_window:i].values
            if len(seq) == pattern_window:
                sequences.append(seq)
                valid_indices.append(self.df.index[i])

        # Generate embeddings
        with torch.no_grad():
            embs = []
            for s in sequences:
                tensor = torch.Tensor(s).unsqueeze(0).to(device)  # Add batch dim
                emb = model(tensor).cpu().numpy()
                embs.append(emb.squeeze())  # Remove singleton dimensions

        # Handle multidimensional embeddings
        if len(embs) > 0 and isinstance(embs[0], np.ndarray):
            # Create multiple columns for multidimensional embeddings
            for i in range(embs[0].shape[0]):
                self.df[f'pattern_emb_{i}'] = np.nan
                self.df.loc[valid_indices, f'pattern_emb_{i}'] = [e[i] for e in embs]
        else:
            # Single dimension case
            self.df['pattern_emb'] = np.nan
            self.df.loc[valid_indices, 'pattern_emb'] = embs

        # Forward fill embeddings
        self.df = self.df.ffill()

        return self

    # 9. Market Memory Features
    def memory_features(self, memory_lag=20):
        """Quantifies market memory effects"""
        def causal_entropy(series):
            hist = series[-memory_lag:]
            return np.exp(-entropy(hist.value_counts(normalize=True)))

        self.df['market_memory'] = self.df['close'].rolling(memory_lag).apply(
            causal_entropy, raw=False)
        return self

    def execute_all(self):
        """Run all feature calculations with default parameters"""
        return (self
                .microstructure_features()
                .auction_features()
                .fractal_features()
                .liquidity_features()
                .regime_features()
                .noise_features()
                .pattern_features()
                .memory_features())
```

```
def volume_weighted_macd(df, fast=12, slow=26, signal=9):
    vw_price = df['close'] * df['volume']
    df['VW_MACD'] = vw_price.ewm(span=fast).mean() - vw_price.ewm(span=slow).mean()
    df['VW_MACD_Signal'] = df['VW_MACD'].ewm(span=signal).mean()
    return df
```

```
def adaptive_volatility(df, window=14):
    true_range = df['high'] - df['low']
    directional_move = abs(df['close'] - df['open']).ewm(span=window).mean()
    df[f'APVR_{window}'] = true_range / (directional_move + 1e-8)
    return df
```

```
from AEDataset import create_dataset_for_autoencoder, Time_series_Dataset
from AEArchitecture import auto_encoder
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

def AE_INDICATORS_OSCILLATORS(df,features = [], window_size = 24,
                  train_size = 0.8, batch_size = 32,epochs = 600, lr = 0.0001,
                  conv1_kernel_num = 9, conv1_kernel_size = 7, conv2_kernel_num = 6,
                   conv2_kernel_size = 7,conv3_kernel_num = 3, conv3_kernel_size = 7,
                   maxpool_kernel_size = 2,max_pool_stride = 2, sanity = False):

  dataset = create_dataset_for_autoencoder(df, features= features, ws = window_size)

  n = int(train_size * len(dataset))
  m = len(dataset) - n
  start_train = n % batch_size
  end_test = m % batch_size

  train_set = dataset[start_train:n]
  if end_test:
    val_set = dataset[n:-end_test]
  else:
    val_set = dataset[n:]


  train_dataset = Time_series_Dataset(train_set)
  val_dataset = Time_series_Dataset(val_set)

  train_dataloader = DataLoader(train_dataset, batch_size = batch_size)
  val_dataloader = DataLoader(val_dataset, batch_size= batch_size)

  if sanity:
    for data_window, label in train_dataloader:
      break
    torch.manual_seed(42)
    model = auto_encoder(channel_dim = len(features), conv1_kernel_num = conv1_kernel_num, conv1_kernel_size = conv1_kernel_size,
                         conv2_kernel_num = conv2_kernel_num, conv2_kernel_size = conv2_kernel_size, conv3_kernel_num = conv3_kernel_num,
                         conv3_kernel_size = conv3_kernel_size ,maxpool_kernel_size = maxpool_kernel_size, max_pool_stride = max_pool_stride,
                         window_size = window_size)

    model.cuda()
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr = lr)

    model.train()

    data_window = data_window.permute(0,2,1)
    label = label.permute(0,2,1)
    for i in range(20000):

      out = model(data_window)
      loss = criterion(out, label)
      print(f'Loss: {loss}')

      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
    print('\nsanity check is done')
    return (True,True,True)

  else:
    print("Training Process...")
    torch.manual_seed(42)
    model = auto_encoder(channel_dim = len(features), conv1_kernel_num = conv1_kernel_num, conv1_kernel_size = conv1_kernel_size,
                         conv2_kernel_num = conv2_kernel_num, conv2_kernel_size = conv2_kernel_size, conv3_kernel_num = conv3_kernel_num,
                         conv3_kernel_size = conv3_kernel_size ,maxpool_kernel_size = maxpool_kernel_size, max_pool_stride = max_pool_stride,
                         window_size = window_size)

    model.cuda()
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr = lr)

    losses = []
    model.train()
    for i in range(epochs):
      for seq,y_train in train_dataloader:
        seq = seq.permute(0,2,1)
        y_train = y_train.permute(0,2,1)

        optimizer.zero_grad()
        y_pred = model(seq)
        loss = criterion(y_pred, y_train)

        loss.backward()
        optimizer.step()
        #scheduler.step()
      losses.append(loss)
      print(f"Epoch {i} Loss: {(loss.item())}")

    print("\nTraining Process has finished")
    print("#############################")

    if train_size != 1.0:
      model.eval()
      print("Evaluate the model...")
      with torch.no_grad():

        for x,y in val_dataloader:

          x = x.permute(0,2,1)
          y = y.permute(0,2,1)
          y_pred = model(x)
          loss = criterion(y_pred, y)
          print(f'Loss: {loss.item()**0.5}')
      print("\nEvaluation has finished")
      print("#############################")
      print("Visualization...")

      for data_window_, label_ in val_dataloader:
        break

      data_window_ = data_window_.permute(0,2,1)
      label_ = label_.permute(0,2,1)
      data_window_ = torch.Tensor.cpu(data_window_)
      label_ = torch.Tensor.cpu(label_)
      pre = model(data_window_.cuda())
      pre = torch.Tensor.cpu(pre)
      fig=plt.figure(figsize=(8, 4))
      fig.add_subplot(1, 3, 1)
      plt.imshow(data_window_[:,:,:3])
      fig.add_subplot(1, 3, 2)
      plt.imshow(label_[:,:,:3])
      fig.add_subplot(1, 3, 3)
      plt.imshow(pre.detach().numpy()[:,:,:3])

    return (model, dataset)
```

#Main Model Architecture

```
import torch.nn as nn
import torch.nn.functional as F
class three_line_network(nn.Module):

  def __init__(self, channel_dim_OHLCV , channel_dim_INDICATORS, channel_dim_OSCILLATORS,
               conv1_kernel_num_OHLCV , conv1_kernel_size_OHLCV, conv2_kernel_num_OHLCV, conv2_kernel_size_OHLCV,
               conv1_kernel_num_IND , conv1_kernel_size_IND, conv2_kernel_num_IND, conv2_kernel_size_IND,
               conv1_kernel_num_OSC , conv1_kernel_size_OSC, conv2_kernel_num_OSC, conv2_kernel_size_OSC,
               maxpool_kernel_size, max_pool_stride, input_size, lstm_hidden_size, lstm_num_layers, batch_size, output_size):


    super().__init__()

    ###OHLCV

    self.channel_dim_OHLCV = channel_dim_OHLCV
    self.conv1_kernel_num_OHLCV = conv1_kernel_num_OHLCV
    self.conv1_kernel_size_OHLCV = conv1_kernel_size_OHLCV
    self.conv2_kernel_num_OHLCV = conv2_kernel_num_OHLCV
    self.conv2_kernel_size_OHLCV = conv2_kernel_size_OHLCV

    ###INDICATORS

    self.channel_dim_INDICATORS = channel_dim_INDICATORS
    self.conv1_kernel_num_INDICATORS = conv1_kernel_num_IND
    self.conv1_kernel_size_INDICATORS = conv1_kernel_size_IND
    self.conv2_kernel_num_INDICATORS = conv2_kernel_num_IND
    self.conv2_kernel_size_INDICATORS = conv2_kernel_size_IND

    ###OSCILLATORS

    self.channel_dim_OSCILLATORS = channel_dim_OSCILLATORS
    self.conv1_kernel_num_OSCILLATORS = conv1_kernel_num_OSC
    self.conv1_kernel_size_OSCILLATORS = conv1_kernel_size_OSC
    self.conv2_kernel_num_OSCILLATORS = conv2_kernel_num_OSC
    self.conv2_kernel_size_OSCILLATORS = conv2_kernel_size_OSC

    ###LSTM

    self.input_size = input_size
    self.lstm_hidden_size = lstm_hidden_size
    self.lstm_num_layers = lstm_num_layers

    ###OTHERS

    self.maxpool_kernel_size = maxpool_kernel_size
    self.max_pool_stride = max_pool_stride
    self.batch_size = batch_size
    self.output_size = output_size



    ##################################################### CONV_OHLCV Path ####################################################

    self.conv1_OHLCV = nn.Conv1d(self.channel_dim_OHLCV, self.conv1_kernel_num_OHLCV, self.conv1_kernel_size_OHLCV, stride=1)
    self.bn1_OHLCV = nn.BatchNorm1d(self.conv1_kernel_num_OHLCV)

    self.conv2_OHLCV = nn.Conv1d(self.conv1_kernel_num_OHLCV, self.conv2_kernel_num_OHLCV, self.conv2_kernel_size_OHLCV, stride=1)
    self.bn2_OHLCV = nn.BatchNorm1d(self.conv2_kernel_num_OHLCV)

    ##################################################### CONV_INDICATORS Path #################################################

    self.conv1_IND = nn.Conv1d(self.channel_dim_INDICATORS, self.conv1_kernel_num_INDICATORS, self.conv1_kernel_size_INDICATORS, stride=1)
    self.bn1_IND = nn.BatchNorm1d(self.conv1_kernel_num_INDICATORS)

    self.conv2_IND = nn.Conv1d(self.conv1_kernel_num_INDICATORS, self.conv2_kernel_num_INDICATORS, self.conv2_kernel_size_INDICATORS, stride=1)
    self.bn2_IND = nn.BatchNorm1d(self.conv2_kernel_num_INDICATORS)

    ##################################################### CONV_OSCILLATOR Path ##################################################

    self.conv1_OSC = nn.Conv1d(self.channel_dim_OSCILLATORS, self.conv1_kernel_num_OSCILLATORS, self.conv1_kernel_size_OSCILLATORS, stride=1)
    self.bn1_OSC = nn.BatchNorm1d(self.conv1_kernel_num_OSCILLATORS)

    self.conv2_OSC = nn.Conv1d(self.conv1_kernel_num_OSCILLATORS, self.conv2_kernel_num_OSCILLATORS, self.conv2_kernel_size_OSCILLATORS, stride=1)
    self.bn2_OSC = nn.BatchNorm1d(self.conv2_kernel_num_OSCILLATORS)


    self.pool = nn.MaxPool1d(kernel_size = self.maxpool_kernel_size, stride = self.max_pool_stride)

    ##################################################### LSTM ###################################################################

    self.lstm = nn.LSTM(input_size=self.input_size, hidden_size=self.lstm_hidden_size, num_layers=self.lstm_num_layers, dropout=0.5)


    self.hidden = (torch.zeros(self.lstm_num_layers, self.batch_size,self.lstm_hidden_size).cuda(),
                   torch.zeros(self.lstm_num_layers, self.batch_size, self.lstm_hidden_size).cuda())


    ##################################################### FC layers ###############################################################

    self.fc1 = nn.Linear(self.lstm_hidden_size,10)
    self.fc2 = nn.Linear(10,self.output_size)




  def forward(self, OHLCV, INDS, OSCS):

    #### OHLCV

    #print(OHLCV.shape)

    X = self.bn1_OHLCV(self.conv1_OHLCV(OHLCV))
    X = F.relu(X)
    X = self.pool(X)

    #print(X.shape)

    X = self.bn2_OHLCV(self.conv2_OHLCV(X))
    X = F.relu(X)

    #print(X.shape)



    #### INDS

    #print(INDS.shape)

    Y = self.bn1_IND(self.conv1_IND(INDS))
    Y = F.relu(Y)
    Y = self.pool(Y)

    #print(Y.shape)

    Y = self.bn2_IND(self.conv2_IND(Y))
    Y = F.relu(Y)

    #print(Y.shape)



    #### OSCS

    #print(OSCS.shape)

    Z = self.bn1_OSC(self.conv1_OSC(OSCS))
    Z = F.relu(Z)
    Z = self.pool(Z)

    #print(Z.shape)

    Z = self.bn2_OSC(self.conv2_OSC(Z))
    Z = F.relu(Z)

    #print(Z.shape)

    convs_output = torch.cat((X,Y,Z), dim=1)


    #print(convs_output.shape)

    '''

    input_size = The number of features of input sequence



    input & (h_0,c_0) -> lstm unit

    input   :  (seq_len, batch, input_size)
    h_0,c_0 :  (num_layers * num_directions, batch, hidden_size)



    lstm unit -> output & (h_n, c_n)

    output  :  (seq_len, batch, num_directions * hidden_size)
    h_n,c_n :  (num_layers * num_directions, batch, hidden_size)

    '''

    lstm_out, self.hidden = self.lstm(convs_output.permute(2,0,1), self.hidden)



    fc = F.relu(self.fc1(lstm_out[-1,:,:]))
    #print(fc.shape)
    fc = self.fc2(fc)
    #print(fc.shape)

    return F.sigmoid(fc)
```

```
def cfm(lower, upper):

  global clipped_y_test, clipped_y_pred, bet_sizing, clipped_ret

  clipped_y_test = np.array([])
  clipped_y_pred = np.array([])
  bet_sizing = np.array([])
  clipped_ret = np.array([])

  for i in range(len(y_pred_)-1):

    if y_pred_[i]<=lower:

      clipped_y_test = np.append(clipped_y_test, y_test_[i])
      clipped_y_pred = np.append(clipped_y_pred, 0)
      bet_sizing = np.append(bet_sizing, abs(y_pred_[i]-0.5))
      clipped_ret = np.append(clipped_ret, ret[i])

    elif y_pred_[i]>=upper:

      clipped_y_test = np.append(clipped_y_test, y_test_[i])
      clipped_y_pred = np.append(clipped_y_pred, 1)
      bet_sizing = np.append(bet_sizing, abs(y_pred_[i]-0.5))
      clipped_ret = np.append(clipped_ret, ret[i])

  cm = confusion_matrix(clipped_y_pred,clipped_y_test)
  make_confusion_matrix(cm, figsize=(10,8), cbar=True)
```
